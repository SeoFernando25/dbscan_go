```python
import pandas as pd
import plotly.graph_objects as go
```

```python
# Load the dataset (clusters_points.csv)
df = pd.read_csv('clusters.csv')
df.head()
```

```python

fig = go.Figure(data=go.Scattergeo(
        lon = df['Longitude'],
        lat = df['Latitude'],
        text = df['Size'],
        marker_color = df['ClusterId'],
        ))

fig.update_layout(
        title = 'DBSCAN Results',
        geo_scope='usa',
    )

fig.show()
```
